```python
#1. Прочитайте csv файл (метод read_csv)
import numpy as np
import pandas as pd

df = pd.read_csv('bestsellers with categories.csv',  delimiter=',', encoding='latin1')
df
```

```python
#2. Виведіть перші 10 рядків (метод head)
dfnew = df.head(10)
print (dfnew)
```

```python
#3. Виведіть розміри датасету (метод shape)
num_rows, num_columns = df.shape
print(f"Кількість рядків: {num_rows}, Кількість стовпчиків: {num_columns}")
print(f"Даний датасет зберігає інформацію про {num_rows} книг")
```

```python
#4. Змініть назви всіх колонок: назва з маленької букви, без пробілів через підкреслення.
df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
df.columns = df.columns.str.replace(' ', '_')    
df
```

```python
#5. Вивести кількість пропусків у кожному стовпчику.
missing_values = df.isnull().sum()

# Визначити, чи є пропуски в будь-якому зі стовпців
has_missing_values = missing_values.any()

print("Кількість пропусків у кожному стовпчику:")
print(missing_values)

if has_missing_values:
    print("Чи є пропуски в будь-якому зі стовпців: Так")
else:
    print("Чи є пропуски в будь-якому зі стовпців: Ні")
```

```python
#6. За допомогою методу unique  перевірте. Які є унікальні жанри в стовпці.
print("Кількість унікальних країн: ", len(df['genre'].unique()))
print("Кількість унікальних країн: ", df['genre'].unique())
```

```python
#7. Визначте максимальну, мінімальну, медіанну ціну за допомогою методів  max, min, mean, median
high_price = df["price"].max()
min_price = df["price"].min()
mean_price = df["price"].mean()
median_price = df["price"].median()

print("Максимальна ціна: ", high_price , "usd")
print("Мінімальна ціна: ", min_price , "usd")
print("Середня ціна: ", mean_price , "usd")
print("Медіанна ціна: ", median_price , "usd")
```

```python
#8. Виконайте наступні завдання з пошуку і сортування даних:
#8.1 Питання: Який рейтинг у датасеті найвищий? 
#8.2 Питання: Скільки книг мають такий рейтинг? 
#8.3 Питання: Яка книга має найбільше відгуків? 
#8.4 Питання: З тих книг, що потрапили до Топ-50 у 2010 році, яка книга найдорожча ?
#8.5 Питання : Скільки книг жанру Fiction потрапили до Топ-50 у 2012 році 
#8.6 Питання : Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або метод isin)? 

#8.1
max_rate = df['user_rating'].max()
print("Найвищий рейтинг: ", max_rate)

#8.2
max_rate =  len(df[df["user_rating"] == max_rate])
print("Найвищий рейтинг усього у:", max_rate, "книг")

#8.3
max_review = df[df["reviews"] == df["reviews"].max()]
print('Книга "', max_review["name"].values[0], '" має найбільшу кількість відгуків.')

#8.4
df_2010_top =  df[df['year'] == 2010].max()
print("Найдорожча книга 2010 року:")
df_2010_top

#8.5
df_2012_num = df[df['genre'] == 'Fiction']
print("Кількість книг 2012 року жанру Fiction становить:", df_2012_num.shape[0])

#8.6
count_df=df.isin({'user_rating': [4.9], 'year': [2010, 2011]})
sum = 0

for i in range(0, count_df.shape[0]):
    for j in range(0, count_df.shape[1]):
        
        if count_df.iat[i, j] == True:
            sum = sum + 1
print(
    "Кількість книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках становить: ",
    sum,
)
```

```python
#9. Вивести максимальну і  мінімальну ціну для жанру Fiction I NonFiction (за допомогою методів  groupby та agg,
#для підрахунку мінімальних та максимальних значень використовуйте max та min). Використовуйте тількі потрібні вам стовпці.

df_NonFiction = df[df["genre"] != "Fiction"]

df_Fiction = df[df["genre"] == "Fiction"]

min_NonFiction = df_NonFiction.groupby('price').min().reset_index()
print("Мінімальна ціна жанру NonFiction:", min_NonFiction.iat[0,0])

min_Fiction = df_Fiction.groupby('price').min().reset_index()
print("Мінімальна ціна жанру Fiction:", min_Fiction.iat[0,0])

max_NonFiction = df_NonFiction.groupby('price').max().reset_index()
print("Максимальна ціна жанру NonFiction:", min_NonFiction.iat[len(min_Fiction)-1,0])

max_Fiction = df_Fiction.groupby('price').max().reset_index()
print("Максимальна ціна жанру Fiction:", max_Fiction.iat[len(max_Fiction)-1,0])
```
