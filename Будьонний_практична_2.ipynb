```python
#4. Створіть масив розміром 5x5, заповніть його випадковими дійсними числами
#в діапазоні від 0 до 1 та знайдіть середнє значення елементів.

mass4 = np.random.rand(5,5)
print("Масив:")
print(mass4)
average_value = np.mean(mass4)
print("    ")
print("Середнє значення:", average_value)
```

```python
#5. Створіть два одновимірних масиви розміром 5, заповніть їх випадковими цілими числами
#в діапазоні від -10 до 10 та виконайте на них поелементні операції додавання, віднімання та множення.

mass5 = np.random.randint(-10, 11, size=(5, 5))
mass6 = np.random.randint(-10, 11, size=(5, 5))

print("Перший масив:")
print(mass5)
print("    ")
print("Другий масив:")
print(mass6)
print("    ")
print("Додавання:")
print(mass5 + mass6)
print("    ")
print("Віднімання:")
print(mass5 - mass6)
print("    ")
print("Множення:")
print(mass5 * mass6)
```

```python
#6. Створіть два вектори розміром 6, заповніть їх довільними дійсними
#числами та знайдіть їх скалярний добуток.

vector1 = np.array([1.1, 4.3, 6.5, 2.8, 2.2, 8.5])
vector2 = np.array([2.8, 1.2, 3.1, 7.3, 3.4, 1.7])

print("Перший вектор:")
print(vector1)
print("    ")
print("Другий вектор:")
print(vector2)

scalar_product = np.dot(vector1, vector2)
print("    ")
print("Скалярний добуток:", scalar_product)
```

```python
#7. Створіть дві матриці розміром 2x3 та 3x2, заповніть їх випадковими цілими числами в
#діапазоні від 1 до 15 та перемножте їх між собою.

mass7 = np.random.randint(0, 16, size=(2, 3))
mass8 = np.random.randint(0, 16, size=(3, 2))
mass9 = mass8.reshape((2,3))

print("Перша матриця:")
print(mass7)
print ("   ")
print("Друга матриця:")
print(mass9)
print ("  ")
print("Множення:")
print (mass7 * mass9)
```

```python
#8. Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні
#від 1 до 12 та знайдіть її обернену матрицю.

mass10 = np.random.randint(0, 13, size=(3, 3))


print("Матриця: ")
print(mass10)
print ("  ")

inv_matrix = np.linalg.inv(mass10)
print("Інвертована матриця: ")
print(inv_matrix)
```

```python
#9. Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в діапазоні
#від -8 до 8 та транспонуйте її.

mass11 = np.random.randint(-8, 9, size=(4, 4))
print("Матриця: ")
print(mass11)
print("  ")

transp = mass11.T
print("Транспонована матриця: ")
print(transp)
```

```python
#10. Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть їх випадковими цілими числами
#в діапазоні від 1 до 10 та знайдіть середнє значення елементів матриці.

mass12 = np.random.randint(1, 11, size=(3, 4))
vector3 = np.random.randint(1, 11, size=4)

print("Матриця: ")
print(mass12)
print ("  ")
print("Вектор: ")
print(vector3)
print ("  ")

print("Середнє значення: ", mass12.mean())
```

```python
#11. Створіть матрицю розміром 2x3 та вектор розміром 3, заповніть матрицю випадковими цілими
#числами в діапазоні від 1 до 10, а вектор - випадковими дійсними числами від 0 до 1,
#і знайдіть добуток вектора на матрицю.

mass13 = np.random.randint(1, 11, size=(2, 3))
vector4 = np.random.random(3)

print("Матриця: ")
print(mass13)
print ("  ")
print("Вектор: ")
print(vector4)
print("  ")
print("Добуток вектора на матрицю: ")
result = np.dot(mass13, vector4)
print(result)
```
