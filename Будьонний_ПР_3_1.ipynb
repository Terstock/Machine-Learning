```python
#9. Визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean)
maxium_views = df["video views"].max()
print("Максимальна кількість переглядів каналу:", maxium_views, "переглядів")

minimum_views = df["video views"].min()
print("Мінімальна кількість переглядів каналу:", minimum_views, "переглядів")

mean_views = df["video views"].mean()
print("Середня кількість переглядів:", mean_views, "переглядів")
```
